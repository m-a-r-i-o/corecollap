# Topological structure of post-core collapse oscillations

Question(s): can the effective dynamics of the post core-collapse oscillations in globular star clusters be described by a low-dimension attractor? If yes, what are its topological properties? Do these properties stay the same if we use different methods to simulate our globular cluster?

Methods: I consider the time series of two properties of the core in a simulated star cluster: core density and velocity dispersion. I extract values separated in time so that the autocorrelation function drops to zero (within significance). I reconstruct the presumed attractor by taking the lagged time series at three different lags (Takens embedding). I calculate the persistence diagrams of the resulting point cloud, and their persistent entropy.
I then randomly shuffle the time series N times and apply the same procedure. I compare the persistent entropy of the real data to the distribution of persistent entropies of the shuffled data. Finally, I redo the whole procedure on data generated from the Rossler attractor rather than on our real data.

Answer: the persistent entropy of our density and velocity dispersion data does not deviate from the distribution of the persistent entropies of the reshuffled data. This suggests that there is no real topological structure in our data at least given the following caveats (hello Mircea). On the other hand the Rossler attractor clearly has a very different persistent entropy with respect to the reshuffled version of itself.

What to do: is this because of MPCDSS? Would a direct N-body simulation show structure? Unfortunately the 10k particle simulation does not even clearly show core collapse happening. Maybe we could get a much bigger simulation from Petar?
Or maybe it's just the coordinates we chose: I bet that with the right transformation of the Rossler attractor and adding sufficient noise we could plug its hole too (but note that the persistent entropy of voids is different too...).


```python
import numpy as np
import pandas as pd
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.tsatools import detrend
from statsmodels.tsa.tsatools import add_lag
from ripser import ripser
from persim import plot_diagrams
from persim.landscapes import plot_landscape_simple
from persim.persistent_entropy import *
np.random.seed(888)
```

```python
cc = pd.read_csv('core_collapse_data.csv', sep = ';') #nbody.csv 
cc.head()
```

### Central density and velocity dispersion as a function of time
Let's plot the central density (mass per unit volume) within a fixed radius around the center of mass of the cluster. 

```python
density_color = '#AC1208'
veldisp_color = '#E3B23C'
box_color = '#756D54' #'#C9B79C'
t_min = 12000
t_max = cc['time'].max() + 50
rho_min = 58
rho_max = 90
sigma_min = 0.3
sigma_max = 0.41
zoom_in_box_t = [t_min, t_min, t_max, t_max, t_min]
zoom_in_box_rho = [rho_min, rho_max, rho_max, rho_min, rho_min]
zoom_in_box_sigma = [sigma_min, sigma_max, sigma_max, sigma_min, sigma_min]
fig, axs = plt.subplots(2,2, figsize=(9, 9), dpi=300)
axs[0,0].plot(cc['time'], cc['rho_c'], color = density_color)
axs[0,0].plot(zoom_in_box_t, zoom_in_box_rho, color = box_color)
axs[0,0].set_ylabel('Central density')
axs[1,0].plot(cc['time'], cc['sigma_c'], color = veldisp_color)
axs[1,0].plot(zoom_in_box_t, zoom_in_box_sigma, color = box_color)
axs[1,0].set_xlabel('t')
axs[1,0].set_ylabel('Central velocity dispersion')
axs[0,1].plot(cc['time'], cc['rho_c'], color = density_color)
axs[0,1].set_xlim([t_min, t_max])
axs[0,1].set_ylim([rho_min, rho_max])
axs[1,1].plot(cc['time'], cc['sigma_c'], color = veldisp_color)
axs[1,1].set_xlim([t_min, t_max])
axs[1,1].set_ylim([sigma_min, sigma_max])
axs[1,1].set_xlabel('t')
plt.show()
```

The density starts at $\approx 1$ ($0.32$ in fact) but climbs up to $\approx 80$, to decrease later on. The moment the maximum density is attained is the core-collapse.
Let's zoom in to see this better.

### Detrend the time series

```python
from statsmodels.nonparametric.smoothers_lowess import lowess
t_around_cc = cc['time'][cc['time'] > t_min]
rho_c_around_cc = cc['rho_c'][cc['time'] > t_min]
sigma_c_around_cc = cc['sigma_c'][cc['time'] > t_min]
detrended_rho_c_around_cc = rho_c_around_cc - lowess(rho_c_around_cc, t_around_cc, return_sorted=False, frac = 0.2)
detrended_sigma_c_around_cc = sigma_c_around_cc - lowess(sigma_c_around_cc, t_around_cc, return_sorted=False, frac = 0.2)

fig, axs = plt.subplots(2,2, figsize=(9, 9), dpi=300)
axs[0,0].plot(t_around_cc, rho_c_around_cc, color = density_color)
axs[0,0].plot(t_around_cc, rho_c_around_cc - detrended_rho_c_around_cc, color = box_color)
axs[0,0].set_ylabel('Central density')
axs[1,0].plot(t_around_cc, sigma_c_around_cc, color = veldisp_color)
axs[1,0].plot(t_around_cc, sigma_c_around_cc - detrended_sigma_c_around_cc, color = box_color)
axs[1,0].set_xlabel('t')
axs[1,0].set_ylabel('Central velocity dispersion')
axs[0,1].plot(t_around_cc, detrended_rho_c_around_cc, color = density_color)
axs[1,1].plot(t_around_cc, detrended_sigma_c_around_cc, color = veldisp_color)
axs[1,1].set_xlabel('t')
plt.show()
```

```python
fig, axs = plt.subplots(2, figsize=(9, 9), dpi=300)
axs[0].set_ylabel("Autocorrelation")
axs[1].set_ylabel("Autocorrelation")
axs[1].set_xlabel("Lag")
_ = plot_acf(detrended_rho_c_around_cc, lags = 2+10*np.arange(80), ax = axs[0], title=None, use_vlines=False, zero=False, auto_ylims=True, color = density_color)
_ = plot_acf(detrended_sigma_c_around_cc, lags = 2+10*np.arange(80), ax = axs[1], title=None, use_vlines=False, zero=False, auto_ylims=True, color = veldisp_color)
```

It looks like the fist lag at which the autocorrelation loses significance is about 500. According to [Ravishanker & Chen 2021](https://wires.onlinelibrary.wiley.com/doi/full/10.1002/wics.1548?casa_token=NjV3TVkgfG4AAAAA%3AWgKxwvpeNn3xHMbwiJl0D0Tz4Cfj5RZjNUtE-3GrvCxEE664HJ0wC1TjyRkxKdGwEz-a803AK_xQee7p) we should choose $\tau \approx 500$ to build our Takens embedding. As for the dimension $d$ we suspect it to be around two or three. Let's choose $4$ to splurge. 

```python
tau_delay = 500
attractor_dimension = 4

selected = np.array([int(x) for x in range(len(detrended_rho_c_around_cc)) if x%tau_delay==0])
subsampled_detrended_rho_c_around_cc = detrended_rho_c_around_cc.to_numpy()[selected]
Nlags_rho = add_lag(subsampled_detrended_rho_c_around_cc, lags=attractor_dimension-1)
subsampled_detrended_sigma_c_around_cc = detrended_sigma_c_around_cc.to_numpy()[selected]
Nlags_sigma = add_lag(subsampled_detrended_sigma_c_around_cc, lags=attractor_dimension-1)
```

```python
if (attractor_dimension > 3):
    print("Attractor dimension is > 3; skipping visualization")

if (attractor_dimension == 3):
    fig = plt.figure(figsize=plt.figaspect(0.5), dpi=300)
    ax = fig.add_subplot(1, 2, 1, projection='3d')
    ax.scatter(Nlags_rho[:,0], Nlags_rho[:,1], Nlags_rho[:,2], color = density_color)
    ax = fig.add_subplot(1, 2, 2, projection='3d')
    ax.scatter(Nlags_sigma[:,0], Nlags_sigma[:,1], Nlags_sigma[:,2], color = veldisp_color)
    plt.show()
    
if (attractor_dimension == 2):
    fig = plt.figure(figsize=plt.figaspect(0.5), dpi=300)
    ax = fig.add_subplot(1, 2, 1)
    ax.plot(Nlags_rho[:,0], Nlags_rho[:,1], color = density_color)
    ax = fig.add_subplot(1, 2, 2)
    ax.plot(Nlags_sigma[:,0], Nlags_sigma[:,1], color = veldisp_color)
    plt.show()
    plt.plot(Nlags_rho[:,0], Nlags_sigma[:,0], color = box_color)
    plt.show()
```

Next step is to compute a persistence diagram on this data (the presumed attractor). We will then compute a persistence diagram on the velocity dispersion (following a similar detrending and subsampling procedure). If the topology of the two data sets is similar, as it should be if by takens theorem they are embedding the same attractor, then we can confirm that core oscillations live near a low-dimensional attractor. Maybe.

```python
diagrams_rho = ripser(Nlags_rho, maxdim=2)['dgms']
plot_diagrams(diagrams_rho, show=True)
```

```python
diagrams_sigma = ripser(Nlags_sigma, maxdim=2)['dgms']
plot_diagrams(diagrams_sigma, show=True)
```

### Null-scenario: if we randomly permute the time series how does the PD look like?

```python
def reshuffled_persistent_entropy():
    permuted_detrended_rho_c_around_cc = np.copy(detrended_rho_c_around_cc.to_numpy())
    np.random.shuffle(permuted_detrended_rho_c_around_cc)
    permuted_detrended_sigma_c_around_cc = np.copy(detrended_sigma_c_around_cc.to_numpy())
    np.random.shuffle(permuted_detrended_sigma_c_around_cc)
    permuted_subsampled_detrended_rho_c_around_cc = permuted_detrended_rho_c_around_cc[selected]
    permuted_subsampled_detrended_sigma_c_around_cc = permuted_detrended_sigma_c_around_cc[selected]
    randomized_Nlags_rho = add_lag(permuted_subsampled_detrended_rho_c_around_cc, lags=attractor_dimension - 1)
    randomized_Nlags_sigma = add_lag(permuted_subsampled_detrended_sigma_c_around_cc, lags=attractor_dimension - 1)
    return np.array([*persistent_entropy(ripser(randomized_Nlags_rho, maxdim=2)['dgms']),
                     *persistent_entropy(ripser(randomized_Nlags_sigma, maxdim=2)['dgms'])])

persistent_entropies = np.array([*persistent_entropy(ripser(Nlags_rho, maxdim=2)['dgms']),
                                      *persistent_entropy(ripser(Nlags_sigma, maxdim=2)['dgms'])]).reshape([1,6])

for i in range(1000):
    persistent_entropies = np.concatenate([persistent_entropies, reshuffled_persistent_entropy().reshape([1,6])], axis=0)
```

```python
shuffle_color = '#B0A3D4'
vline_height = 100

plt.hist(persistent_entropies[1:,0], color = shuffle_color)
plt.vlines(persistent_entropies[0,0], 0, vline_height, color = density_color)
plt.show()

plt.hist(persistent_entropies[1:,1], color = shuffle_color)
plt.vlines(persistent_entropies[0,1], 0, vline_height, color = density_color)
plt.show()

plt.hist(persistent_entropies[1:,2], color = shuffle_color)
plt.vlines(persistent_entropies[0,2], 0, vline_height, color = density_color)
plt.show()

plt.hist(persistent_entropies[1:,3], color = shuffle_color)
plt.vlines(persistent_entropies[0,3], 0, vline_height, color = veldisp_color)
plt.show()

plt.hist(persistent_entropies[1:,4], color = shuffle_color)
plt.vlines(persistent_entropies[0,4], 0, vline_height, color = veldisp_color)
plt.show()

plt.hist(persistent_entropies[1:,5], color = shuffle_color)
plt.vlines(persistent_entropies[0,5], 0, vline_height, color = veldisp_color)
plt.show()
```

Whoa... the persistent entropy of the actual data (red and yellow bars) does not seem much different if at all from the distribution of the reshuffled data (purple histogram)!

#### So the fluctuations around the density and velocity dispersion trends are indeed just noise!

Let's double check this by analyzing in the same fashion something that is definitely NOT noise.

```python
rossler_color = '#5BC0EB'

def rossler_eq(y, t, a, b, c):
    x, y, z = y
    dxdt = -y - z
    dydt = x + a*y
    dzdt = b + z*(x - c)
    return [dxdt, dydt, dzdt]

a = 0.2
b = 0.2
c = 5.7
y0 = [1, 1, 1]
t = np.linspace(0, 1000, len(permuted_detrended_rho_c_around_cc))

sol = odeint(rossler_eq, y0, t, args=(a, b, c))

x = sol[:, 0]
y = sol[:, 1]
z = sol[:, 2]

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(z, y, x, color = rossler_color)
plt.show()

plt.plot(t,x, color = rossler_color)
plt.show()
```

```python
def reshuffled_persistent_entropy_rossler():
    permuted_rossler = np.copy(x)
    np.random.shuffle(permuted_rossler)
    permuted_subsampled_rossler = permuted_rossler[selected]
    randomized_Nlags_rossler = add_lag(permuted_subsampled_rossler, lags=attractor_dimension - 1)
    return persistent_entropy(ripser(randomized_Nlags_rossler, maxdim=2)['dgms'])

subsampled_rossler = x[selected]
Nlags_rossler = add_lag(subsampled_rossler, lags=attractor_dimension - 1)

persistent_entropies_rossler = persistent_entropy(ripser(Nlags_rossler, maxdim=2)['dgms']).reshape([1,3])
                                
for i in range(1000):
    persistent_entropies_rossler = np.concatenate([persistent_entropies_rossler, 
                                           reshuffled_persistent_entropy_rossler().reshape([1,3])], axis=0)

```

```python
plt.hist(persistent_entropies_rossler[1:,0], color = shuffle_color)
plt.vlines(persistent_entropies_rossler[0,0], 0, vline_height, color = rossler_color)
plt.show()

plt.hist(persistent_entropies_rossler[1:,1], color = shuffle_color)
plt.vlines(persistent_entropies_rossler[0,1], 0, vline_height, color = rossler_color)
plt.show()

plt.hist(persistent_entropies_rossler[1:,2], color = shuffle_color)
plt.vlines(persistent_entropies_rossler[0,2], 0, vline_height, color = rossler_color)
plt.show()
```

### This pretty much settles the case: our data (MPCDSS core density and velocity dispersion) have persistent entropy that comes from the same distribution as the reshuffled version of themselves. The Rossler attractor does not.
